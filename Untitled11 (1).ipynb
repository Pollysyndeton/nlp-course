```python
import pandas as pd

data = pd.read_csv('IMDB Dataset.csv')
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words("english"))
```

```python
import re
texts = []
for review in data['review']:
  a = re.sub(r'https?://\S+', '', review)
  b = re.sub(r'[@#]\w+', '', a)
  c = re.sub(r'\+7\s\d{3}\s\d{3}\s\d{2}\s\d{2}', '', b)
  d = re.sub(r'\+7\d{10}', '', c)
  e = re.sub(r'\W+', ' ', d) # удаление знаков препинания
  f = re.sub(r'[0-9]+', ' ', e)
  texts.append(f)
```

```python
texts = texts[::2]
```

```python
cleaned_texts = []
for text in texts:
    words = text.split()
    cleaned_words = []
    for word in words:
        lower_word = word.lower()
        if lower_word not in stop_words:
            cleaned_words.append(lower_word)
    cleaned_text = ' '.join(cleaned_words)
    cleaned_texts.append(cleaned_text)
```

```python
print(cleaned_texts[0])
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()

tokenizer.fit_on_texts(cleaned_texts)

sequences = tokenizer.texts_to_sequences(cleaned_texts) # токенизируем все данные
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_len = 200
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post', truncating='post')
#добавляем нули чтобы не возникло проблем с слоями в модели. добавляем нули в конец и обрезаем тоже с конца
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['sentiment'] = le.fit_transform(data['sentiment']) #преобразование позитива и негатива в 0 и 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, data['sentiment'], test_size=0.2, random_state=42)
```

```python
from tensorflow.keras import layers, models# Выстраиваем модель
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, Dense, Flatten
# Параметры модели
vocab_size = len(tokenizer.word_index) + 1
model = Sequential([
    Embedding(vocab_size, 128, input_length=X_train.shape[1]),
    Conv1D(128, 5, activation='relu'), #размер ядра - 5, количество фильтров - 128
    MaxPooling1D(2), #размер окна - 2
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

# Оцениваем модель
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc:.4f}")

# Отрисовка графика прогресса обучения
plt.figure(figsize=(12, 4))

# Отрисовка аккуратности (точности) модели
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Отрисовка значений функции потерь
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
